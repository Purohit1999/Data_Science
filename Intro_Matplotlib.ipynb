```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: 10 years
years = np.arange(2010, 2020)
kohli_runs = [800, 900, 950, 1100, 1200, 1000, 1300, 1250, 1400, 1500]
rohit_runs = [500, 700, 800, 1000, 1150, 1100, 1200, 1350, 1450, 1600]
sehwag_runs = [1000, 950, 870, 920, 1000, 970, 880, 900, 850, 800]

# Apply one custom style
plt.style.use('seaborn-v0_8-darkgrid')


# Plot each player's runs
plt.plot(years, kohli_runs, label='Kohli', color='blue', linestyle='--', marker='o')
plt.plot(years, rohit_runs, label='Rohit Sharma', color='green', linestyle='-', marker='s')
plt.plot(years, sehwag_runs, label='Sehwag', color='red', linestyle=':', marker='^')

# Labels and title
plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs Comparison: Kohli, Rohit Sharma, Sehwag (2010-2019)')

# Legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
