```python
import pandas as pd
import numpy as np
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=4,n_classes=2,random_state=42)
```

```python
X.shape
```

```python
y.shape
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
base_model=DecisionTreeClassifier(max_depth=1)
```

```python
model=AdaBoostClassifier(estimator=base_model,n_estimators=200,random_state=42)
```

```python
model.fit(Xtrain,ytrain)
```

```python
print('training',model.score(Xtrain,ytrain))
print('testing',model.score(Xtest,ytest))
```

```python
ypred=model.predict(Xtest)
accuracy_score(ytest,ypred)
```

Adaboost Regressor implementation


```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=1000,n_features=4,n_informative=1,noise=2,random_state=42)
```

```python
X.shape
```

```python
y.shape
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
base_model=DecisionTreeRegressor(max_depth=1)
```

```python
model=AdaBoostRegressor(estimator=base_model,n_estimators=200,random_state=42)
```

```python
model.fit(Xtrain,ytrain)
```

```python
print('training',model.score(Xtrain,ytrain))
print('testing',model.score(Xtest,ytest))
```
