```python
import requests
from bs4 import BeautifulSoup
import time

# Target URL
url = "https://books.toscrape.com"

# User-Agent header to mimic a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

try:
    # Always use timeout to avoid hanging
    response = requests.get(url, headers=headers, timeout=5)

    # Check if request was successful
    if response.status_code == 200:
        print("✅ Page fetched successfully!")
        
        # Parse HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract and display book titles
        books = soup.find_all('h3')
        print("\n📚 Book Titles on the Page:\n")
        for book in books:
            title = book.a["title"]
            print("- " + title)

        # OPTIONAL: Pause to avoid hitting server too frequently
        time.sleep(1)

    else:
        print(f"⚠️ Request returned with status code {response.status_code}")

except requests.exceptions.RequestException as e:
    print(f"❌ Connection error occurred: {e}")
```

```python
import requests
from bs4 import BeautifulSoup
import os

# Target URL
url = "https://books.toscrape.com"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

# Create folders
folder_path = os.path.join(os.getcwd(), "Requests", "html")
os.makedirs(folder_path, exist_ok=True)
file_path = os.path.join(folder_path, "books.html")

try:
    response = requests.get(url, headers=headers, timeout=5)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, "html.parser")
    books = soup.find_all("h3")

    # Build clean HTML structure using BeautifulSoup
    new_soup = BeautifulSoup("<html><head><title>Book Titles</title></head><body></body></html>", "html.parser")
    body = new_soup.body
    body.append(new_soup.new_tag("h1"))
    body.h1.string = "Book Titles from books.toscrape.com"

    ul = new_soup.new_tag("ul")
    for book in books:
        title = book.a["title"]
        li = new_soup.new_tag("li")
        li.string = title
        ul.append(li)
    body.append(ul)

    # Write prettified HTML to file
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(new_soup.prettify())

    print("✅ Prettified HTML saved at:", file_path)

except requests.exceptions.RequestException as e:
    print(f"❌ Error occurred: {e}")
```

```python
import os
import time
import requests
from bs4 import BeautifulSoup

# Set User-Agent header to mimic a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

all_books = []  # This list will store all scraped book titles

# Loop through pages 1 to 50
for page in range(1, 51):
    # Construct the URL depending on the page number
    if page == 1:
        url = "https://books.toscrape.com/"
    else:
        url = f"https://books.toscrape.com/catalogue/page-{page}.html"
        
    try:
        response = requests.get(url, headers=headers, timeout=5)
        response.raise_for_status()  # Raise an error for unsuccessful responses

        soup = BeautifulSoup(response.text, "html.parser")
        books = soup.find_all("h3")
        
        # If no books found, exit the loop early
        if not books:
            print(f"Page {page} has no book listings; stopping further requests.")
            break

        for book in books:
            title = book.a["title"]
            all_books.append(title)

        print(f"✅ Fetched {len(books)} books from page {page}")

    except requests.exceptions.RequestException as e:
        print(f"❌ Error fetching page {page}: {e}")
        
    # Pause for 1 second between pages to be polite to the server
    time.sleep(1)

# --- Generate the HTML file with all book titles ---

# Create the basic HTML structure using BeautifulSoup
new_soup = BeautifulSoup("<html><head><title>Book Titles (Pages 1-50)</title></head><body></body></html>", "html.parser")
body = new_soup.body

# Create and add a header
header_tag = new_soup.new_tag("h1")
header_tag.string = "Book Titles from pages 1 to 50 of books.toscrape.com"
body.append(header_tag)

# Create a list (ul) and add each book title as a list item (li)
ul = new_soup.new_tag("ul")
for title in all_books:
    li = new_soup.new_tag("li")
    li.string = title
    ul.append(li)

body.append(ul)

# Create the folder path where the HTML file will be stored.
folder_path = os.path.join(os.getcwd(), "Requests", "html")
os.makedirs(folder_path, exist_ok=True)

# Define the output file path and save the prettified HTML content.
file_path = os.path.join(folder_path, "books_pages_1_50.html")
with open(file_path, "w", encoding="utf-8") as file:
    file.write(new_soup.prettify())

print("✅ Prettified HTML saved at:", file_path)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Headers to mimic a browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

# List to store scraped book data
items = []

# Map class rating text to star value
rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

# Loop through pages 1 to 50
for page in range(1, 51):
    # Page 1 is slightly different in URL
    if page == 1:
        url = "https://books.toscrape.com/"
    else:
        url = f"https://books.toscrape.com/catalogue/page-{page}.html"

    try:
        response = requests.get(url, headers=headers, timeout=5)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, "html.parser")

        # Each book is inside <article class="product_pod">
        books = soup.find_all("article", class_="product_pod")

        if not books:
            print(f"⚠️ No books found on page {page}. Stopping.")
            break

        for book in books:
            title = book.h3.a["title"]
            price = book.find("p", class_="price_color").text.strip().replace("£", "")
            rating_class = book.p.get("class")[1]  # Second class name is rating
            rating = rating_map.get(rating_class, 0)

            items.append([title, price, rating])

        print(f"✅ Scraped page {page} successfully with {len(books)} books.")
        time.sleep(1)  # Be polite to server

    except requests.exceptions.RequestException as e:
        print(f"❌ Failed to scrape page {page}: {e}")
        break

# Create DataFrame
df = pd.DataFrame(items, columns=["Book Title", "Price", "Rating"])
print("📊 DataFrame created with", len(df), "entries.")

# Optionally export to CSV
df.to_csv("books_data.csv", index=False)
print("💾 Saved to books_data.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Headers to mimic a browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

# List to store scraped book data
items = []

# Map class rating text to star value
rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

# Loop through pages 1 to 50
for page in range(1, 51):
    # Page 1 has a different URL format
    url = "https://books.toscrape.com/" if page == 1 else f"https://books.toscrape.com/catalogue/page-{page}.html"

    try:
        response = requests.get(url, headers=headers, timeout=5)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, "html.parser")
        books = soup.find_all("article", class_="product_pod")

        if not books:
            print(f"⚠️ No books found on page {page}. Stopping.")
            break

        for book in books:
            title = book.h3.a["title"]
            # Remove currency symbol and convert to float
            price = float(book.find("p", class_="price_color").text.strip().replace("£", ""))
            rating_class = book.p.get("class")[1]
            rating = rating_map.get(rating_class, 0)

            items.append([title, price, rating])

        print(f"✅ Scraped page {page} successfully with {len(books)} books.")
        time.sleep(1)

    except requests.exceptions.RequestException as e:
        print(f"❌ Failed to scrape page {page}: {e}")
        break

# Create DataFrame
df = pd.DataFrame(items, columns=["Book Title", "Price", "Rating"])
print("📊 DataFrame created with", len(df), "entries.")

# Export to CSV (safe encoding, clean price)
df.to_csv("books_data.csv", index=False, encoding="utf-8")
print("💾 CSV saved as books_data.csv")

# Export to Excel (requires openpyxl)
try:
    df.to_excel("books_data.xlsx", index=False)
    print("📘 Excel file saved as books_data.xlsx")
except ImportError:
    print("⚠️ Install 'openpyxl' to enable Excel export: pip install openpyxl")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import re  # For cleaning price strings

# Headers to mimic a browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

# List to store scraped book data
items = []

# Map class rating text to star value
rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

# Loop through pages 1 to 50
for page in range(1, 51):
    # Page 1 has a different URL format
    url = "https://books.toscrape.com/" if page == 1 else f"https://books.toscrape.com/catalogue/page-{page}.html"

    try:
        response = requests.get(url, headers=headers, timeout=5)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, "html.parser")
        books = soup.find_all("article", class_="product_pod")

        if not books:
            print(f"⚠️ No books found on page {page}. Stopping.")
            break

        for book in books:
            title = book.h3.a["title"]

            # Clean price string using regex to remove non-numeric characters
            raw_price = book.find("p", class_="price_color").text.strip()
            clean_price = re.sub(r"[^\d.]", "", raw_price)
            price = float(clean_price)

            # Extract and convert rating
            rating_class = book.p.get("class")[1]
            rating = rating_map.get(rating_class, 0)

            items.append([title, price, rating])

        print(f"✅ Scraped page {page} successfully with {len(books)} books.")
        time.sleep(1)

    except requests.exceptions.RequestException as e:
        print(f"❌ Failed to scrape page {page}: {e}")
        break

# Create DataFrame
df = pd.DataFrame(items, columns=["Book Title", "Price", "Rating"])
print("📊 DataFrame created with", len(df), "entries.")

# Export to CSV (safe encoding, clean price)
df.to_csv("books_data.csv", index=False, encoding="utf-8")
print("💾 CSV saved as books_data.csv")

# Export to Excel (requires openpyxl)
try:
    df.to_excel("books_data.xlsx", index=False)
    print("📘 Excel file saved as books_data.xlsx")
except ImportError:
    print("⚠️ Install 'openpyxl' to enable Excel export: pip install openpyxl")
```
