```python
#import Seaborn
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set theme
sns.set_theme(style='darkgrid')  # Try 'whitegrid', 'ticks', 'white', 'dark'

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=y, label='sin(x)', color='blue', linewidth=2.5)
sns.lineplot(x=x, y=y2, label='cos(x)', color='green', linewidth=2.5, linestyle='--')

# Add titles and labels
plt.title('Trigonometric Functions', fontsize=16)
plt.xlabel('X values (0 to 10)', fontsize=12)
plt.ylabel('Function Output', fontsize=12)

# Add annotation
plt.text(3, 0.1, 'sin(x) peaks here', fontsize=10, color='blue')
plt.axvline(x=1.57, color='gray', linestyle=':', label='π/2')

# Show legend
plt.legend(title='Functions', loc='upper right')

# Display
plt.tight_layout()
plt.show()


```
