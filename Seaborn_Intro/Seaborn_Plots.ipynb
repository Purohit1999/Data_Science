```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Set overall theme
sns.set_theme(style='whitegrid')

# --- 1. LINE PLOT ---
tips = sns.load_dataset('tips')
plt.figure(figsize=(8, 4))
sns.lineplot(x='total_bill', y='tip', data=tips, color='blue')
plt.title('Line Plot: Tip vs Total Bill', fontsize=14)
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.tight_layout()
plt.show()

# --- 2. SCATTER PLOT ---
plt.figure(figsize=(8, 4))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='time', style='sex', s=100)
plt.title('Scatter Plot: Tip vs Bill by Time and Gender', fontsize=14)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.tight_layout()
plt.show()

# --- 3. BAR PLOT ---
plt.figure(figsize=(8, 4))
sns.barplot(x='day', y='total_bill', data=tips, palette='pastel', estimator='mean', errorbar='sd')
plt.title('Average Total Bill by Day', fontsize=14)
plt.xlabel('Day of Week')
plt.ylabel('Avg Bill Amount ($)')
plt.tight_layout()
plt.show()

# --- 4. BOX PLOT ---
plt.figure(figsize=(8, 4))
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set2')
plt.title('Boxplot: Total Bill Distribution per Day', fontsize=14)
plt.tight_layout()
plt.show()

# --- 5. HEATMAP (from correlation matrix) ---
flights = sns.load_dataset('flights')
pivot_table = flights.pivot(index='month', columns='year', values='passengers')

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='YlGnBu', linewidths=.5)
plt.title('Heatmap of Monthly Airline Passengers (1949-1960)', fontsize=14)
plt.tight_layout()
plt.show()

# --- 6. CUSTOM DATAFRAME: SCATTER BY GENDER ---
df = pd.DataFrame({
    "age": [22, 25, 47, 52, 46, 56, 55, 60, 34, 43],
    "salary": [25000, 27000, 52000, 60000, 58000, 62000, 61000, 65000, 38000, 45000],
    "gender": ["M", "F", "F", "M", "F", "F", "F", "M", "F", "M"]
})

plt.figure(figsize=(8, 4))
sns.scatterplot(x="age", y="salary", hue="gender", style="gender", data=df, s=100, palette="husl")
plt.title("Salary vs Age by Gender", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Salary ($)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set theme
sns.set_theme(style='whitegrid')

# Simulated dataset
np.random.seed(42)
n = 50
departments = ['Dev', 'Data', 'QA', 'HR']
genders = ['M', 'F']

df = pd.DataFrame({
    'Age': np.random.randint(22, 50, n),
    'Experience': np.random.randint(1, 20, n),
    'Salary': np.random.randint(35000, 120000, n),
    'Department': np.random.choice(departments, n),
    'Gender': np.random.choice(genders, n)
})

# Derived column (e.g., seniority)
df['Seniority'] = ['Senior' if x > 10 else 'Junior' for x in df['Experience']]

# ----- 1. SCATTERPLOT: Age vs Salary -----
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Salary', hue='Gender', style='Department', size='Experience', sizes=(20, 200), palette='muted')
plt.title('Age vs Salary by Gender, Department, and Experience')
plt.tight_layout()
plt.show()

# ----- 2. BARPLOT: Avg Salary per Department -----
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Department', y='Salary', palette='Set2', estimator=np.mean, errorbar='sd')
plt.title('Average Salary by Department')
plt.tight_layout()
plt.show()

# ----- 3. BOX PLOT: Salary Distribution by Gender -----
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Gender', y='Salary', palette='coolwarm')
plt.title('Salary Distribution by Gender')
plt.tight_layout()
plt.show()

# ----- 4. VIOLIN PLOT: Experience by Department -----
plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x='Department', y='Experience', hue='Gender', split=True, palette='husl')
plt.title('Experience Distribution per Department by Gender')
plt.tight_layout()
plt.show()

# ----- 5. HEATMAP: Correlation Matrix -----
plt.figure(figsize=(8, 5))
corr = df[['Age', 'Experience', 'Salary']].corr()
sns.heatmap(corr, annot=True, cmap='Blues', fmt='.2f')
plt.title('Correlation Matrix: Age, Experience & Salary')
plt.tight_layout()
plt.show()
```
