```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
years = ['2001', '2002', '2003', '2004', '2005']
sachin_runs = [1200, 1100, 1300, 1250, 1000]
kohli_runs = [800, 900, 1000, 950, 1050]
sehwag_runs = [950, 980, 1020, 900, 920]

# Bar width and positions
x = np.arange(len(years))
bar_width = 0.25

# Side-by-side bar chart
plt.figure(figsize=(10, 6))
plt.bar(x - bar_width, sachin_runs, width=bar_width, label='Sachin', color='skyblue')
plt.bar(x, kohli_runs, width=bar_width, label='Kohli', color='green')
plt.bar(x + bar_width, sehwag_runs, width=bar_width, label='Sehwag', color='orange')

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Runs Comparison Over 5 Selected Years")
plt.xticks(x, years)
plt.legend()
plt.tight_layout()
plt.show()

# Horizontal bar chart for total debut 5-year runs
total_runs = {
    'Sachin': sum(sachin_runs),
    'Kohli': sum(kohli_runs),
    'Sehwag': sum(sehwag_runs)
}

plt.figure(figsize=(8, 4))
plt.barh(list(total_runs.keys()), list(total_runs.values()), color=['skyblue', 'green', 'orange'])
plt.xlabel("Total Runs")
plt.title("Total Runs in First 5 Years of Career")
plt.tight_layout()
plt.show()
```
