```python
# Setup
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Sample data
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]

# 1. Basic Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(study_hours, exam_scores)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
# 2. Adding Color and Size
sizes = [score * 2 for score in exam_scores]
colors = ['red' if score < 60 else 'green' for score in exam_scores]

plt.figure(figsize=(8, 5))
plt.scatter(study_hours, exam_scores, s=sizes, c=colors)
plt.title('Colored & Sized Scatter Plot')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
# 3. Using a Colormap
scores_normalized = np.array(exam_scores)

plt.figure(figsize=(8, 5))
scatter = plt.scatter(study_hours, exam_scores, c=scores_normalized, cmap='viridis')
plt.colorbar(scatter, label='Score')
plt.title('Scatter Plot with Colormap')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
# 4. Adding Annotations
plt.figure(figsize=(8, 5))
plt.scatter(study_hours, exam_scores)
for i in range(len(study_hours)):
    plt.annotate(f'Student {i+1}', (study_hours[i], exam_scores[i]))
plt.title('Scatter Plot with Annotations')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
# 5. Multiple Groups in One Plot
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.figure(figsize=(8, 5))
plt.scatter(class_a_hours, class_a_scores, label='Class A', color='blue')
plt.scatter(class_b_hours, class_b_scores, label='Class B', color='orange')
plt.title('Scatter Plot: Class A vs Class B')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
sleep_hours = [4, 5, 6, 6.5, 7, 7.5, 8, 8.5, 9, 10]
productivity_score = [50, 55, 60, 63, 68, 72, 78, 80, 83, 85]
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(sleep_hours, productivity_score)
plt.title('Sleep Hours vs Productivity Score')
plt.xlabel('Hours of Sleep')
plt.ylabel('Productivity Score')
plt.grid(True)
plt.show()
```

```python
# Size: scaled based on productivity score
sizes = [score * 2 for score in productivity_score]
# Color: red if < 70, else green
colors = ['red' if score < 70 else 'green' for score in productivity_score]

plt.figure(figsize=(8, 5))
plt.scatter(sleep_hours, productivity_score, s=sizes, c=colors)
plt.title('Colored & Sized Scatter Plot: Sleep vs Productivity')
plt.xlabel('Hours of Sleep')
plt.ylabel('Productivity Score')
plt.grid(True)
plt.show()
```

```python
scores_normalized = np.array(productivity_score)

plt.figure(figsize=(8, 5))
scatter = plt.scatter(sleep_hours, productivity_score, c=scores_normalized, cmap='plasma')
plt.colorbar(scatter, label='Productivity Score')
plt.title('Colormap: Sleep vs Productivity')
plt.xlabel('Hours of Sleep')
plt.ylabel('Productivity Score')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(sleep_hours, productivity_score)
for i in range(len(sleep_hours)):
    plt.annotate(f'Person {i+1}', (sleep_hours[i], productivity_score[i]))
plt.title('Annotated Plot: Sleep vs Productivity')
plt.xlabel('Hours of Sleep')
plt.ylabel('Productivity Score')
plt.grid(True)
plt.show()
```
