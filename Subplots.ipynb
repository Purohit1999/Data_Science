```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
sleep_hours = [4, 5, 6, 6.5, 7, 7.5, 8, 8.5, 9, 10]
productivity_score = [50, 55, 60, 63, 68, 72, 78, 80, 83, 85]
sizes = [score * 2 for score in productivity_score]
colors = ['red' if score < 70 else 'green' for score in productivity_score]
scores_normalized = np.array(productivity_score)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
```

```python
# 1. Basic Scatter Plot
axs[0, 0].scatter(sleep_hours, productivity_score)
axs[0, 0].set_title('Basic Scatter Plot')
axs[0, 0].set_xlabel('Sleep Hours')
axs[0, 0].set_ylabel('Productivity')
axs[0, 0].grid(True)

# 2. Colored & Sized Plot
axs[0, 1].scatter(sleep_hours, productivity_score, s=sizes, c=colors)
axs[0, 1].set_title('Colored & Sized Points')
axs[0, 1].set_xlabel('Sleep Hours')
axs[0, 1].set_ylabel('Productivity')
axs[0, 1].grid(True)
```

```python
# 3. Colormap Plot
scatter = axs[1, 0].scatter(sleep_hours, productivity_score, c=scores_normalized, cmap='plasma')
axs[1, 0].set_title('Colormap Scatter')
axs[1, 0].set_xlabel('Sleep Hours')
axs[1, 0].set_ylabel('Productivity')
axs[1, 0].grid(True)
fig.colorbar(scatter, ax=axs[1, 0], label='Score')
```

```python
# 4. Annotated Plot
axs[1, 1].scatter(sleep_hours, productivity_score)
for i in range(len(sleep_hours)):
    axs[1, 1].annotate(f'P{i+1}', (sleep_hours[i], productivity_score[i]))
axs[1, 1].set_title('Scatter with Annotations')
axs[1, 1].set_xlabel('Sleep Hours')
axs[1, 1].set_ylabel('Productivity')
axs[1, 1].grid(True)
```

```python
# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]           # Double of x
y2 = [i ** 2 for i in x]         # Square of x
y3 = [i ** 0.5 for i in x]       # Square root of x
y4 = [10 - i for i in x]         # Reverse x from 10

y_data = [y1, y2, y3, y4]
titles = ['x * 2', 'x squared', 'sqrt(x)', '10 - x']
colors = ['blue', 'green', 'purple', 'orange']
```

```python
# Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
axs = axs.ravel()  # Flatten 2D array for easy indexing

for i in range(4):
    axs[i].plot(x, y_data[i], marker='o', color=colors[i])
    axs[i].set_title(titles[i], fontsize=12)
    axs[i].set_xlabel('x')
    axs[i].set_ylabel('y')
    axs[i].grid(True)
```

```python
# Add overall title
fig.suptitle('Math Functions Comparison using Subplots', fontsize=16)

# Adjust layout and spacing
fig.tight_layout()
fig.subplots_adjust(top=0.90)  # Shift to make room for suptitle

# Optional: Save the figure
fig.savefig("math_functions_subplot.png")

# Show plot
plt.show()
```
