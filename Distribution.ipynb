```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('winequalityN.csv')

# Display basic info
print("Initial Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())

# Drop rows with missing values (or you can fill them)
df.dropna(inplace=True)

# Confirm changes
print("\nAfter cleaning, shape:", df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Plot distribution of selected features
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar']

plt.figure(figsize=(14, 8))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.histplot(df[feature], kde=True, bins=30)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x=df[feature])
    plt.title(f'Boxplot of {feature}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Wine Features')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='quality', hue='type')
plt.title('Wine Quality Count by Type')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example: joint distribution between 'alcohol' and 'quality'
sns.jointplot(data=df, x='alcohol', y='quality', kind='scatter', hue='type')  # 'type' is optional for color
plt.show()
```

```python
sns.jointplot(data=df, x='citric acid', y='residual sugar', kind='kde', fill=True)
plt.show()
```

```python
sns.jointplot(data=df, x='fixed acidity', y='volatile acidity', kind='hex', color='purple')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example with selected features
sns.pairplot(df[['alcohol', 'quality', 'residual sugar', 'citric acid']])
plt.show()
```

```python
sns.pairplot(df[['alcohol', 'quality', 'residual sugar', 'citric acid', 'type']], hue='type')
plt.show()
```

```python
sns.pairplot(df[['alcohol', 'quality', 'residual sugar', 'citric acid']], diag_kind='kde')
plt.show()
```

# 📊 Wine Quality Data Analysis - README Summary

## 📁 Dataset

**File:** `winequalityN.csv`
This dataset contains physicochemical properties and quality ratings for **red and white wine** samples.

---

## 🔍 Exploratory Data Analysis Summary

### ✅ **1. Data Cleaning**

* Missing values were found and **dropped** for a clean analysis.
* Final dataset shape after cleaning: `(rows, columns)` *(actual count depends on your file)*

---

### 📈 **2. Distribution Plots**

Plotted for key features such as:

* `fixed acidity`, `volatile acidity`, `citric acid`, `residual sugar`

**Key Insights:**

* Most features are **right-skewed**, especially `residual sugar`.
* `citric acid` has a near-normal distribution but with a mild tail.

---

### 📊 **3. Boxplots (Outlier Detection)**

* Outliers are present in features like `residual sugar`, `chlorides`, and `free sulfur dioxide`.
* Helps in considering feature scaling or transformations before ML modeling.

---

### 🔥 **4. Correlation Heatmap**

* `alcohol` is **positively correlated** with `quality`.
* `volatile acidity` shows a **negative correlation** with `quality`.
* `density` is highly correlated with `residual sugar` and `fixed acidity`.
* Most other features have **low to moderate** correlations.

---

### 🔁 **5. Joint Plot Examples**

* **`alcohol` vs `quality`**:

  * Higher alcohol levels generally correlate with better quality ratings.
* **`citric acid` vs `residual sugar`**:

  * Slight linear relation; denser in low-moderate values.

---

### 🔗 **6. Pair Plot (Multivariate Relationship)**

Features Plotted: `alcohol`, `quality`, `residual sugar`, `citric acid`

**Insights:**

* Quality 6 is the **most common rating**.
* Clear **clustering by wine type** when using `hue='type'`.
* Alcohol is a visually distinguishing feature between wine types and quality ratings.

---

## 📌 Recommendations for Modeling

* Consider **feature scaling** due to presence of outliers and skewed distributions.
* Use top correlated features like `alcohol`, `volatile acidity`, and `sulphates` in predictive models.
* Wine `type` can be an important categorical variable for stratified modeling.
