```python
import matplotlib.pyplot as plt

# Data
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

# Create vertical bar chart
plt.figure(figsize=(6, 4))
plt.bar(players, runs, color="#99eb95")  # light green bar

# Add labels on top of each bar
for i in range(len(players)):
    plt.text(i, runs[i] + 50, str(runs[i]), ha='center', fontsize=10)

# Label and title
plt.ylabel("Runs")
plt.title("Runs Scored by Players")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]
colors = ['#4CAF50', '#FF9800', '#2196F3']  # Distinct colors for each player

# Create horizontal bar chart
plt.figure(figsize=(8, 4))
bars = plt.barh(players, runs, color=colors)

# Add value labels next to the bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 30, bar.get_y() + bar.get_height() / 2,
             str(width), va='center', fontsize=10)

# Chart formatting
plt.xlabel("Runs")
plt.title("Runs Scored by Players (Horizontal Bar Chart)")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()

plt.show()
```
